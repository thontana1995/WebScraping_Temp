```python
import pandas as pd
```

```python
df = pd.read_csv('Temperature.csv', encoding = 'utf8')
```

```python
df.set_index('Province', inplace = True)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.sort_index(inplace = True)
```

```python
df['Temperature'] = df['Temperature'].map(lambda x: x.rstrip('°C'))
```

```python
df['Temperature'] = pd.to_numeric(df['Temperature'], downcast='integer')
```

df['Time'] = df.Date.str[25:-3]

```python
df2 = df['Date'].str.split(' ',expand=True)
```

```python
df2[1] = df2[1].apply(lambda x: '01' if 'มกราคม' in x else x)
df2[1] = df2[1].apply(lambda x: '02' if 'กุมภาพันธ์' in x else x)
df2[1] = df2[1].apply(lambda x: '03' if 'มีนาคม' in x else x)
df2[1] = df2[1].apply(lambda x: '04' if 'เมษายน' in x else x)
df2[1] = df2[1].apply(lambda x: '05' if 'พฤษภาคม' in x else x)
df2[1] = df2[1].apply(lambda x: '06' if 'มิถุนายน' in x else x)
df2[1] = df2[1].apply(lambda x: '07' if 'กรกฎาคม' in x else x)
df2[1] = df2[1].apply(lambda x: '08' if 'สิงหาคม' in x else x)
df2[1] = df2[1].apply(lambda x: '09' if 'กันยายน' in x else x)
df2[1] = df2[1].apply(lambda x: '10' if 'ตุลาคม' in x else x)
df2[1] = df2[1].apply(lambda x: '11' if 'พฤศจิกายน' in x else x)
df2[1] = df2[1].apply(lambda x: '12' if 'ธันวาคม' in x else x)
```

```python
df2[2] = df2[2].astype(int) - 543
df2[2] = df2[2].astype(str)
df2 = df2.drop(columns=[3, 4, 6])
```

```python
df2['Datetime'] = df2.apply(lambda row: '-'.join(row.values.astype(str)), axis=1)
```

```python
df2 = df2.drop(columns=[0,1,2,5])
```

```python
result = pd.concat([df, df2], axis=1).reindex(df.index)
```

```python
result = result.drop(columns = 'Date')
```

```python
result['Datetime'] = pd.to_datetime(result['Datetime'])
```
